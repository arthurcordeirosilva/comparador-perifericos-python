```python
import json
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

#  Carregar produtos
def carregar_produtos():
    with open("produtos.json", "r") as f:
        return json.load(f)

#  Filtrar produtos
def filtrar_produtos(produtos, tipo, orcamento):
    filtrados = []

    for p in produtos:
        if p["tipo"] == tipo:
            if orcamento == "baixo" and p["preco"] <= 150:
                filtrados.append(p)
            elif orcamento == "medio" and 150 < p["preco"] <= 300:
                filtrados.append(p)
            elif orcamento == "alto" and p["preco"] > 300:
                filtrados.append(p)

    return filtrados

#  Sistema de recomendação (score)
def calcular_score(produto):
    preco_score = 10 - (produto["preco"] / 50)
    qualidade_score = produto["qualidade"] * 2

    return preco_score + qualidade_score

#  Ordenar por melhor custo-benefício
def recomendar(produtos):
    return sorted(produtos, key=calcular_score, reverse=True)

#  Interface simples (terminal)
def main():
    produtos = carregar_produtos()

    print("=== COMPARADOR DE PERIFÉRICOS ===")

    tipo_widget = widgets.Dropdown(
        options=['gamer', 'escritorio'],
        description='Tipo:',
        disabled=False,
    )

    orcamento_widget = widgets.Dropdown(
        options=['baixo', 'medio', 'alto'],
        description='Orçamento:',
        disabled=False,
    )

    output_widget = widgets.Output()

    def on_button_clicked(b):
        with output_widget:
            clear_output()
            tipo = tipo_widget.value
            orcamento = orcamento_widget.value

            filtrados = filtrar_produtos(produtos, tipo, orcamento)

            if not filtrados:
                print("Nenhum produto encontrado.")
                return

            recomendados = recomendar(filtrados)

            print("\n🔝 Melhores opções:\n")

            for p in recomendados:
                print(f"Nome: {p['nome']}")
                print(f"Preço: R${p['preco']}")
                print(f"Qualidade: {p['qualidade']}")
                print("-" * 20)

            media = sum(p["preco"] for p in filtrados) / len(filtrados)

            print(f"\n Média de preço: R${media:.2f}\n")

           # 📊 Criar médias por categoria
            categorias = {"baixo": [], "medio": [], "alto": []}

            for p in produtos:
                if p["tipo"] == tipo:
                    if p["preco"] <= 150:
                        categorias["baixo"].append(p["preco"])
                    elif p["preco"] <= 300:
                        categorias["medio"].append(p["preco"])
                    else:
                        categorias["alto"].append(p["preco"])

            medias = []
            labels = []

            for cat in ["baixo", "medio", "alto"]:
                if categorias[cat]:
                    media_cat = sum(categorias[cat]) / len(categorias[cat])
                    medias.append(media_cat)
                    labels.append(cat)

            # 📈 Plot (sem definir cor!)
            plt.figure(figsize=(8, 5))
            plt.bar(labels, medias)

            plt.xlabel("Categoria de Orçamento")
            plt.ylabel("Preço Médio (R$)")
            plt.title("Média de Preços por Categoria")

            # Mostrar valores em cima das barras
            for i, v in enumerate(medias):
                plt.text(i, v + 5, f"R${v:.0f}", ha='center')

            plt.tight_layout()
            plt.show()

    button = widgets.Button(description="Recomendar")
    button.on_click(on_button_clicked)

    display(tipo_widget, orcamento_widget, button, output_widget)

#  Executar
if __name__ == "__main__":
    main()
```
